```python
import pandas as pd
import numpy as py



```

```python
#Read csv file
csv_file = "CDC_data.csv"
cdc_df = pd.read_csv(csv_file)
cdc_df.head()
```

```python
#only data from Question "Percent of adults aged 18 years and older who have obesity"

cdc_df_filtered = cdc_df.loc[cdc_df["Question"] == "Percent of adults aged 18 years and older who have obesity"]

```

```python
#filter for only 2014 to match other data set
cdc_df_filtered2 = cdc_df_filtered.loc[cdc_df["YearStart"] == 2014]
cdc_df_filtered2.info()
```

```python
#dropna from Gender column
cdc_df_filtered3 = cdc_df_filtered2
cdc_df_filtered3.dropna(subset = ["Gender"], inplace=True)
cdc_df_filtered3.info()
```

```python
cdc_df_filtered3.head()
```

```python
#Retrieve relevant columns
cdc_df_filtered4 = cdc_df_filtered3[['LocationAbbr', 'Data_Value_Alt', 'Gender']]
```

```python
#Rename Columns
cdc_df_clean = cdc_df_filtered4.rename(columns={
    'LocationAbbr': 'stateid',
    'Data_Value_Alt': 'percentage(obese)',
    'Gender': 'gender'
})
```

```python
#save to a csv file
cdc_df_clean.to_csv("cdc_data_clean.csv", index=True)
```
