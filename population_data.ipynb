```python
import pandas as pd
import numpy as py



```

```python
#Read csv file
csv_file = "census_population_data.csv"
population_df = pd.read_csv(csv_file)
population_df.head()
```

```python
#Retrieve relevant columns
pop_remove = population_df[['NAME', 'CENSUS2010POP', 'POPESTIMATE2014']]
pop_remove.info()
```

```python
#Rename Columns
population_df_clean = pop_remove.rename(columns={
    'NAME': 'state',
    'CENSUS2010POP': 'totalpopulation(census 2010)',
    'POPESTIMATE2014': 'population(est 2014)'
})
```

```python
#save to a csv file
population_df_clean.to_csv("population_df_clean.csv", index=True)
```
